```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

Замініть у таблиці значення "—" на значення NaN

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'
```

```python
df = pd.read_html(url, decimal=',', thousands='.', match='Крим')[2]
```

Замініть у таблиці значення "—" на значення NaN

```python
df.replace('—', np.nan, inplace=True)
df
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
df = df.astype(dict.fromkeys(['2014', '2019'], np.float64))
df.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
df.isnull().sum() / df.isnull().sum().sum()
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df_new = df.drop([27])
```

```python
df_new
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df_new = df_new.fillna({'1950': df_new['1950'].mean(), '1960': df_new['1960'].mean(), '1970': df_new['1970'].mean(), '1990': df_new['1990'].mean(), '2000': df_new['2000'].mean(), '2012': df_new['2012'].mean(), '2014': df_new['2014'].mean(), '2019': df_new['2019'].mean()})
df_new
```

```python
df_new.plot()
plt.title('Коефіцієнт народжуваності в регіонах України (1950—2019)')
plt.xlabel("Регіон")
plt.ylabel('Коефіцієнт народжуваності')
plt.show()
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
df_new[['Регіон', '2019']][df_new['2019'] > df_new['2019'].mean()]
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
df_new[['Регіон', '2014']].max()
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
df_new = df_new.set_index(df_new['Регіон'])
df_new.drop(['Регіон'], axis=1)
```

```python
df_new['2019'].plot(kind='bar')
plt.title('Народжуваність по регіонах у 2019 році')
plt.show()
```

```python
df_new['2019'].plot(kind='barh')
plt.title('Народжуваність по регіонах у 2019 році')
plt.ylabel('Регіон')
plt.grid()
```
